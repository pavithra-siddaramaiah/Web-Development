Text manipulation:

so far we have seen text manipulation using .innerHTML 
The problem with this is it will display everything inside the tags
ex:  <h1><strong>Hello</strong></h1>
here we we request document.querySelector("h1").innerHTML;
// <strong>Hello</strong>

Incase when we want the content only not any tags between we for for textContent
document.querySelector("h1").textContent; // Hello

To change the content:
document.querySelector("h1").textContent = "Good Bye"; //Good Bye
make sure to enclose the content in " "

Higher order functions

    return num1+num2;

    return num1-num2;
}
function multiply(num1,num2){
    return num1*num2;
}
function divide(num1, num2){

function calculator(num1, num2, operator){
    return operator(num1,num2);
