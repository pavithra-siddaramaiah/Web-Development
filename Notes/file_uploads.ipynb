File uploads
=============
if we want to upload a file to the database then express does not support this by itself
But we have a npm library called multer using with we can upload the files teh to teh database

const upload = multer({
    dest: 'avatars'
 })
 //here we are assigning the dest as avatars means a folder will be created, named as avatars and the file we send from the postamn will be saved here
  
 router.post('/users/me/avatar', upload.single('avatar'), (req, res) => {
    res.send()
 } )
 //whatever we write inside the single() should be a string and same value we type in key value pair in postman, key- string, value-image

 with the help of this code, you can upload a image in the postaman, 
 in the postman previously for the json data we were going to body ->raw and adding the json data but now will select the body -> form-data and upload the image

Add error handling function
===========================
router.post('/users/me/avatar', upload.single('avatar'), (req, res) => {
    res.send()
 }, (error, req, res, next) => {
    res.status(400).send({error: error.message})
 } )

Uploading the avatar
=====================

Till now what were doing is saving the avatars to local avatars folder which was the destination. Now we will remove that and add the authentication and update the model to use the avatar Buffer and update the router as follows
router.post('/users/me/avatar', auth, upload.single('avatar'), async (req, res) => {
    req.user.avatar = req.file.buffer
    await req.user.save()
    res.send()
    
 }, (error, req, res, next) => {
    res.status(400).send({error: error.message})
 } )



Deleting the avatar
==================
router.delete('/users/me/avatar', auth, async(req, res) =>{
    req.user.avatar = undefined
    await req.user.save()
    res.send('Profile picture successfully deleted')
 })
