Manipulating the styles with jQuery

consider the example of above image where we are using .css("Color", "red"); property to change hte color of the h1.

This can be done by writing the styles in the css like..

    backgroung-color: red;
}
then creating this class in index.js file like
$("h1").addClass("chnaging-color") if we want to add another class, then we can do as below:
$("h1").addClass("changing-color adding-padding") just by adding a space between the classes, inside the same double quotes.

similar to .hide(); we can also use .show(), .toggle(), .fadeOut(), .fadeIn(), fadeToggle(), slideUp(), slideDown();

We can create our own animation using the animate function, the only diadvantage of this function is we can only include the numbber values like margin, padding, opacity whose values are defined with numbers.

})
