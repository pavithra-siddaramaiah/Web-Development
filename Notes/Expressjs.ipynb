HTTP is needed to communicate between 2 devices or we can say from client side to the server side

localhost : making our own computer as a server.

to check the ports in mac => sudo lsof -i -P -n | grep LISTEN
for windows: netstat -ano | finder "LISTENING"



GET - > Request resource from the server.

Each  time when we make changes in the code, we have to stop the server by giving ctrl C and run the code again by node index.js

User do not accept to do this all the time so the solution for this is "nodemon"

this will restart the server automatically by running nodemon index.js

inordert to install this globally we do => npm i -g nodemon



In the code app.get method / means the destination, 
/ corresponds to localhost:3000/ => home page
/about corresponds to localhost:3000/about
/contact corresponds to localhost:3000/contact
